## Exploratory Data Analysis(EDA) of the given dataset

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn.preprocessing as preprocessing 
```

```python
df=pd.read_excel('Course Recommendation System.xlsx')
df=df.drop(['ID','Start time','Completion time','Email','Name','Name(not mandatory)\n' ], axis=1)
df.head()
```

```python
df['mean'] = df.mean(axis=1,skipna=True,numeric_only=True)
df.head()
```

```python
df.shape
```

```python
sns.set(rc={'figure.figsize':(65,8)})
sns.barplot(y=df['mean'],x=[i for i in range(0,df.shape[0])])
df.drop(axis="columns",labels=["mean"],inplace=True)
df
```

```python
mean=df.mean(axis=0,skipna=True,numeric_only=True)
figure, axes = plt.subplots(1, 5, sharex=True,figsize=(105,10))
figure.suptitle('avg rating vs subject')

for i in range(5):
    sns.barplot(y=mean[i:(i+5)],x=[df.columns[i] for i in range(i,(i+5))],ax=axes[i])
    axes[i].set_title('5 subjects')
    for bars in axes[i].containers:
        print(bars.)
        axes[i].bar_label(bars, fmt='%.1f')
# i=5
# sns.barplot(y=mean[i-5:i],x=[df.columns[i] for i in range(i-5,i)],ax=axes[0])
# axes[0].set_title('first 5 subjects')
# i+=5
# sns.barplot(y=mean[i-5:i],x=[df.columns[i] for i in range(i-5,i)],ax=axes[1])
# axes[1].set_title('second 5 subjects')
# i+=5
# sns.barplot(y=mean[i-5:i],x=[df.columns[i] for i in range(i-5,i)],ax=axes[2])
# axes[1].set_title('second 5 subjects')
```
