## Exploratory Data Analysis(EDA) of the given dataset

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn.preprocessing as preprocessing 
```

```python
df=pd.read_excel('Course Recommendation System.xlsx')
df=df.drop(['ID','Start time','Completion time','Email','Name','Name(not mandatory)\n' ], axis=1)
df.head()
```

```python
df['mean'] = df.mean(axis=1,skipna=True,numeric_only=True)
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data (150 users with random mean values)
mean_values = df['mean'].to_numpy()

# Calculate the overall mean of all the users
overall_mean = np.mean(mean_values)

fig = plt.figure(figsize=(8, 6))
# Plot the histogram with the mean values on the x-axis
plt.hist(mean_values, bins=15, alpha=0.65)

# Add a vertical line at the overall mean value
plt.axvline(x=overall_mean, color='red')

# Add labels and a title
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.title('Histogram of Mean Values for 162 Users')

# Show the plot
plt.show()
```

```python
df_core = df[['Data Structures and Algorithms', 'Computer Architecture\n',
       'Discrete Mathematics\n', 'Economics', 'Programming-2']]
df_ai = df[['Machine Learning\n', 'Mathematics For Machine Learning\n',
       'Visual Recognition\n', 'Natural Language Processing\n',
       'Reinforcement Learning']]
df_tscd = df[['Software Production Engineering\n',
       'Computer Graphics\n', 'Data Visualization\n',
       'Programming Languages\n', 'Cloud Computing\n']]
df_vlsi = df[['Digital CMOS VLSI Design ', 'System design with FPGA ', 'ASIC design ',
       'VLSI Architecture Design\n',
       'High level synthesis and optimization of Digital Circuits ']]
df_dt = df[['Digital Sociology ', 'Privacy in the Digital Age',
       'Technology Ethics and AI ', 'Techno-economics of networks ',
       'The Web and the Mind ']]
```

```python
mean_core = df_core.mean(skipna=True,numeric_only=True)
mean_ai = df_ai.mean(skipna=True,numeric_only=True)
mean_tscd = df_tscd.mean(skipna=True,numeric_only=True)
mean_vlsi = df_vlsi.mean(skipna=True,numeric_only=True)
mean_dt = df_dt.mean(skipna=True,numeric_only=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample Pandas Series
# Create a bar graph using Matplotlib
fig = plt.figure(figsize=(8, 6))
mean_dt.plot.bar()

# Add labels and a title
plt.xlabel('Subjects')
plt.ylabel('Mean')
plt.title('Average Rating for DigiSoc Subjects')
plt.xticks(rotation=15)
# Show the plot
plt.show()
```
